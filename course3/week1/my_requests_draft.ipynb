```python
# Make a Request for getting friends list

token = "c3656bf6c3656bf6c3656bf6f5c30fe069cc365c3656bf69fbc16ec0a33e2cc73b2c44c"
user_id = id_

friends_request = requests.get(f'https://api.vk.com/method/friends.get?v=5.71&access_token={token}&user_id={user_id}&fields=bdate')
friends_request_json = friends_request.json()
friends_response = friends_request_json["response"]
friends_items = friends_response["items"]
friends_items
```

```python
import time
```

```python
current_year = time.localtime().tm_year
current_year
```

```python
id_yearofbirth = {item["id"]: int(item["bdate"].split(".")[2]) 
                  for item in friends_items if ("bdate" in item) and (len(item["bdate"].split(".")) == 3)}
id_yearofbirth
```

```python
id_age = {item: current_year - id_yearofbirth[item] for item in id_yearofbirth}
id_age
```

```python
ages_list = list(id_age.values())
ages_list
```

```python
from collections import Counter
```

```python
counter = Counter(ages_list)
```

```python
counted = counter.most_common()
counted
```

```python
sorted_counted1 = sorted(counted, key=lambda x: x[0], reverse=False)
sorted_counted2 = sorted(sorted_counted1, key=lambda x: x[1], reverse=True)
sorted_counted2
```

```python
sorted_counted = sorted(counted, key=lambda x: (x[1], -x[0]), reverse=True)
sorted_counted
```
